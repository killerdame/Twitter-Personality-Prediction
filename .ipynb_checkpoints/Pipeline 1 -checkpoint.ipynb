```python
def cleaned_text(X):
    all_words = [rb_words, gr_words, dt_words, bo_words, mm_words, ow_words, em_words, kk_words, kp_words, nc_words]
    words = X.replace('\(', '').replace('\)', '').replace('@', '').replace('https://.?', '')
    words = re.sub(r"http\S+", "", words)
    corpus = re.sub("['\''%'':''#'')''('';''\n''/''.''!''*''-''+''$'',''?''~''&''@''``''-''=''--''|''<''>']", "", words)
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(corpus.lower())
    filtered_text = [w for w in word_tokens if not w in stop_words]
    for p in all_words:
        return {p: filtered_text}
```
